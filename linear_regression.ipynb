```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression          # import linear regression model
from sklearn.model_selection import train_test_split       # create train data and test data
from sklearn.metrics import mean_squared_error, r2_score   # check  regression model's accuarcy
```

```python
df = pd.read_csv(r"D:\App download\database\auto-mpg.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
features = df[['weight', 'model year']]
target = df['mpg'].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()       # uselinearRegression Model
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)   #  use MSE to check the accuarcy
r2 = r2_score(y_test, y_pred)              #  use r2 to check the accuarcy

print(f"Root_mean_square_error (MSE): {mse}")
print(f"R²: {r2}")
```

The y_test range is between 5 and 45, and an MSE of 9.44 indicates that the model's performance is not very good.

R2[0,1] more trend 1 more better, 0.8 means model is good.

perhaps some outlier caused MSE not accuarcy

```python
plt.scatter(x_test['model year'].values, y_test, color='blue', label='actual')
plt.scatter(x_test['model year'].values, y_pred, color='red', label='pridect')
plt.xlabel('Model year')
plt.ylabel('MPG')
plt.title('Linear Regression')
plt.legend()
plt.show()
```

```python
plt.scatter(x_test['weight'].values, y_test, color='blue', label='actual')
plt.scatter(x_test['weight'].values, y_pred, color='red', label='pridect')
plt.xlabel('Weight')
plt.ylabel('MPG')
plt.title('Linear Regression')
plt.legend()
plt.show()
```

```python
plt.scatter(y_test, y_pred, color='blue', label='Predicted vs Actual')

min_val = min(min(y_test), min(y_pred))
max_val = max(max(y_test), max(y_pred))
plt.plot([min_val, max_val], [min_val, max_val], color='green', label='Ideal Fit')

plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
plt.legend()

plt.show()
```
